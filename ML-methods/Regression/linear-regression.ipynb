## What Problem It Solves

Linear Regression solves regression problems. The goal is to predict a continuous numerical value


based on one or more input features.

##### Examples:


Student marks from study hours


House price from size


Salary from experience

##### Output is always continuous (numbers)

## How It Works (Basic Idea)

The "Linear" in Linear Regression means it tries to draw a straight line through your data points.

This line is the "model."

#### Equation:

##### y = mx + c

|Symbol |	Meaning|
|-------|----------|
|y	    |predicted output (e.g., Marks)|
|x	    |input  (e.g., Hours Studied)|
|m	     | slope (How much does y change when x increases by 1?)|
|c	      |intercept (The value of y when x is 0)|



# The 5 Essential Lines for ANY Regression Model

from sklearn.linear_model import LinearRegression

### 2. Create the model (initialize it)
model = LinearRegression()
