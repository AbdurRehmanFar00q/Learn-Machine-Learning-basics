INDEXING IN NUMPY


Indexing in 1D Array

Write a Python program to access elements of a 1D NumPy array using indexing.

```python
# Import NumPy library
import numpy as np  
# Create a 1D array for indexing examples
arr1 = np.array([1,2,3,4,5])
# Print the original array
print("Original array:", arr1)
# Access and print element at index 1 (second element)
print("Element at index 1:", arr1[1])
# Access and print element at index 0 (first element)
print("Element at index 0:", arr1[0])
# Access and print the last element using negative indexing
print("Last element:", arr1[-1])
```

Write a Python program to access elements of a 2D NumPy array using indexing.

SLICING IN NUMPY


Slicing in 1D Array

Write a Python program to slice elements from a 1D NumPy array

```python
import numpy as np
# Create a 1D array for slicing examples        
arr1 = np.array([1,2,3,4,5])
# Print the original array  
print("Original array:", arr1)
# Slice the array from index 0 to 3 (exclusive)
print("Sliced array (0:3):", arr1[0:3])
# Slice the array from index 2 to the end
print("Sliced array (2:):", arr1[2:])
# Slice the array from the beginning to index 3 (exclusive)
print("Sliced array (:3):", arr1[:3])
# Slice the array with a step of 2
print("Sliced array with step 2 (0:5:2):", arr1[0:5:2])
```

 Slicing in 2D Array

Write a Python program to slice rows and columns from a 2D NumPy array.


```python
import numpy as np
# Create a 2D array for slicing examples
arr2 = np.array([[1,2,3,4,5], [5,6,7,8,9]])
# Print the original 2D array   
print("Original array:\n", arr2)
# Slice the first row (row index 0)
print("Sliced first row (0:1):\n", arr2[0:1])
# Slice the second row (row index 1)
print("Sliced second row (1:2):\n", arr2[1:2])
# Slice the first two columns of the first row
print("Sliced first two columns of the first row (0:1, 0:2):\n", arr2[0:1, 0:2])
# Slice the last three columns of the second row
print("Sliced last three columns of the second row (1:2, 2:5):\n", arr2[1:2, 2:5])
```

Mathematical Opreations

Create a 1D NumPy array and find max, min, sum, and mean, standard deviation

```python
import numpy as np
# Create a 1D array for slicing examples    
arr1 = np.array([1,2,3,4,5])
# Print the original array
print("Original array:", arr1)
# Access and print the maximum value in the array
print("Maximum value in the array:", arr1.max())
# Access and print the minimum value in the array
print("Minimum value in the array:", arr1.min())
# Access and print the sum of all values in the array
print("Sum of all values in the array:", arr1.sum())
# Access and print the mean of all values in the array
print("Mean of all values in the array:", arr1.mean())
# Access and print the standard deviation of all values in the array
print("Standard deviation of all values in the array:", arr1.std())
```

Create two 2D arrays and Perform addition, subtraction, multiplication, matrix multiplication, and division elementwise.

```python
import numpy as np
# Create two 2D arrays for slicing examples
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
# Print the original arrays
print("Array 1:\n", arr1)
print("Array 2:\n", arr2)
# Perform element-wise addition of the two arrays   
print("Element-wise addition of arr1 and arr2:\n", arr1 + arr2)
# Perform element-wise multiplication of the two arrays
print("Element-wise multiplication of arr1 and arr2:\n", arr1 * arr2)
# Perform matrix multiplication of the two arrays   
print("Matrix multiplication of arr1 and arr2:\n", np.dot(arr1, arr2))
# perform element-wise division of the two arrays
print("Element-wise division of arr1 by arr2:\n", arr1 / arr2)
```

Stacking in NumPy

📌 Types of Stacking in NumPy

We mainly use:

vstack() → Vertical stacking,
 hstack() → Horizontal stacking,
  stack() → Add new dimension,
   dstack() → Depth stacking (3D)

```python
import numpy as np

a = np.array([1,2])
b = np.array([3,4])

print("Original Arrays:")
print(a)
print(b)

# Vertical Stack
print("\nVertical Stack:")
print(np.vstack((a,b)))

# Horizontal Stack
print("\nHorizontal Stack:")
print(np.hstack((a,b)))

# Stack with new axis
print("\nStack axis=0:")
print(np.stack((a,b), axis=0))

print("\nStack axis=1:")
print(np.stack((a,b), axis=1))

# Depth stack
a2 = np.array([[1,2],[3,4]])
b2 = np.array([[5,6],[7,8]])

print("\nDepth Stack:")
print(np.dstack((a2,b2)))
```

Broadcasting

What is Broadcasting?

👉 Broadcasting = performing operations on arrays of different shapes

Instead of manually matching sizes, NumPy automatically expands arrays.

```python
import numpy as np

# Example 1: Scalar
arr = np.array([1,2,3])
print("Scalar + Array:", arr + 5)

# Example 2: Row broadcasting
a = np.array([[1,2,3],
              [4,5,6]])
b = np.array([10,20,30])
print("\nRow Broadcasting:\n", a + b)

# Example 3: Column broadcasting
c = np.array([[1],[2],[3]])
d = np.array([10,20,30])
print("\nColumn Broadcasting:\n", c + d)
```
