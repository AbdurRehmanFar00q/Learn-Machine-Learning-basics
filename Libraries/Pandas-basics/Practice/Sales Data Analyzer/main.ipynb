Sales Data Analyzer

```python
import pandas as pd 

data = pd.read_csv("sales-data.csv")

print("Sales data \n", data)
```

```python
total_sales = data["Sales"].sum() # Total Sales 
print("Total sales are ", total_sales)
```

```python
best = data.loc[data["Sales"].idxmax()] # Finding best product
print("Best product is \n", best)
```

```python
grouped = data.groupby("Category")["Sales"].sum()  # group by category
print("Sales by category ", grouped)
```
