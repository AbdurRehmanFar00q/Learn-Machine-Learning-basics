Student Result Analyzer

```python
import pandas as pd 

data = pd.read_csv("studentdata.csv")

print("Student data ", data)
```

```python
data["Average"] = data[['Math' , 'Physics', 'Chemistry']].mean(axis=1) # Finding average marks for each student
print("Average marks are \n", data)  
```

```python
topper = data.loc[data["Average"].idxmax()] # Finding Topper student
print("Topper \n ", topper)
```

```python
data["Status"] = data["Average"].apply(lambda x: "Pass" if x>=60  else "Fail" )  # Pass/Fail

print("Student data with status \n", data)    
```

```python
data.to_csv("studentdata.csv")  # Saving the data to csv file

print(' Final data \n', data)
```
