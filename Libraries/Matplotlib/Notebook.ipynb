Line Plot (MOST BASIC)

```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt 

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 20]

plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")        
plt.show()
```

Scatter Plot (VERY IMPORTANT FOR ML)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]


plt.hist(x)
plt.title("Histogram")
plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 30, 25, 15, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

Multiple Plots

```python
import matplotlib.pyplot as plt

x = [1,2,3]

plt.plot(x, [10,20,30])
plt.plot(x, [15,25,35])

plt.show()
```

## 🎨 Common Styling Functions

| Function | Use |
|----------|-----|
| color="red" | Change color |
| linewidth=2 | Line thickness |
| linestyle="--" | Dashed line |
| marker="o" | Add points |
| markersize=8 | Size of marker |
| alpha=0.5 | Transparency |
| label="Data" | Add legend label |
| plt.legend() | Show legend |
| plt.grid(True) | Show grid |
| plt.figure(figsize=(6,4)) | Set figure size |

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y, color="blue", linestyle="--", marker="o", linewidth=2)

plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.grid(True)
plt.show()
```

## 📊 Bar Chart Functions

| Function | Use |
|----------|-----|
| color="green" | Bar color |
| width=0.5 | Bar width |
| edgecolor="black" | Border color |

```python
import matplotlib.pyplot as plt

categories = ["A","B","C"]
values = [10,20,15]

plt.bar(categories, values, color="green", edgecolor="black")

plt.title("Bar Chart")
plt.show()
```

## 📊 Scatter Plot Functions

| Function | Use |
|----------|-----|
| color="red" | Point color |
| s=100 | Size of points |
| alpha=0.5 | Transparency |

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.scatter(x, y, color="red", s=100, alpha=0.7)

plt.title("Scatter Plot")
plt.show()
```

## 📊 Histogram Functions

| Function | Use |
|----------|-----|
| bins=5 | Number of bars |
| color="purple" | Bar color |
| edgecolor="black" | Border |

```python
import matplotlib.pyplot as plt

data = [10,20,20,30,30,30,40]

plt.hist(data, bins=5, color="purple", edgecolor="black")

plt.title("Histogram")
plt.show()
```

## 🥧 Pie Chart Functions

| Function | Use |
|----------|-----|
| labels=[] | Names |
| colors=[] | Slice colors |
| autopct="%1.1f%%" | Show percentage |
| startangle=90 | Rotate chart |

```python
import matplotlib.pyplot as plt

labels = ["A","B","C"]
sizes = [30,40,30]

plt.pie(sizes, labels=labels, autopct="%1.5f%%", startangle=90)

plt.title("Pie Chart")
plt.show()
```

Multiple Graphs + Legend

```python
import matplotlib.pyplot as plt

x = [1,2,3]

plt.plot(x, [10,20,30], label="Line 1")
plt.plot(x, [15,25,35], label="Line 2")

plt.legend()
plt.show()
```
